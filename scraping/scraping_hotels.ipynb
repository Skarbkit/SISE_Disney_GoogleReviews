```python
def replace_all(text, dic):
    for i, j in dic.items():
        text = text.replace(i, j)
    return text
```

```python
dic_comment = {'<div class="X4nL7d"><div><div class="dA5Vzb"><span class="uTU5Ac">':' ',
    '</span><span>':' ',
    '</span></div><div class="dA5Vzb"><span class="uTU5Ac">':' ',
    '</span></div><div class="dA5Vzb"><span class="uTU5Ac">':' ',
    '</span>':'',
    '<div class="X4nL7d"><div><span class="MoXxzc">':' ',
    '</span><div><div class="dA5Vzb"><span class="uTU5Ac">':' ',
    '<div><div class="dA5Vzb"><span class="uTU5Ac">':' ',
    '</div>':''}
```

```python
def find_between( s, first, last ):
    try:
        start = s.index( first ) + len( first )
        if (last != 'end'):
            end = s.index(last)
        else :
            end = len(s)
        
        return s[start:end]
    except ValueError:
        return ""
```

```python
# Récupère seulement les éléments qui nous intéresse et transforme les résultats en dataframe
def get_review_to_df(result_set):
    rev_dict = {'Review ID' : [],
        'Review Rate': [],
        'Review Time': [],
        'Review Text' : [],
        'Review Voyage Type':[],
        'Review Room':[],
        'Review Service':[],
        'Review Emplacement':[],
        'Review Visite' : []}
    for result in result_set:
        #if (result.find('div', class_='TJUuge')):
        #   result.find('div', class_='TJUuge').click()
        review_rate = int(result.find('div', class_="GDWaad").text[0])

        brut_id = result.find('a',class_ ='DHIhE QB2Jof')
        review_id = find_between(str(brut_id), 'target="_blank">', '</a>')

        
        brut_time = result.find('span',class_='iUtr1 CQYfx')
        review_time = find_between(str(brut_time), '<span class="iUtr1 CQYfx">', 'sur <span class="YhR3n">')

        room_grade = ''
        serv_grade = ''
        empl_grade = ''
        review_tvoyage = ''
        if (result.find('div', class_="X4nL7d")):
            clean_more = replace_all(str(result.find('div', class_="X4nL7d")), dic_comment)
            room_grade = find_between(clean_more, 'Chambres', 'Service').replace(' ','').split(',')[0]
            serv_grade = find_between(clean_more, 'Service', 'Emplacement').replace(' ','').split(',')[0]
            empl_grade = find_between(clean_more, 'Emplacement', 'end').replace(' ','').split(',')[0]
            review_tvoyage = find_between(clean_more, 'Type de voyage', 'Chambre')

            
        

        if (result.find('div', class_="STQFb eoY5cb")):
            brut_text = str(result.find('div', class_="STQFb eoY5cb"))
            review_text = find_between(brut_text, '<div class="K7oBsc"><div><span>','</span></div></div>')


        if (result.find('div', class_='DmVtKb')):
            review_visite = result.find('div', class_='DmVtKb').text
        

        rev_dict['Review ID'].append(review_id)
        rev_dict['Review Rate'].append(review_rate)
        rev_dict['Review Time'].append(review_time)
        rev_dict['Review Text'].append(review_text)
        rev_dict['Review Voyage Type'].append(review_tvoyage)
        rev_dict['Review Room'].append(room_grade)
        rev_dict['Review Service'].append(serv_grade)
        rev_dict['Review Emplacement'].append(empl_grade)
        rev_dict['Review Visite'].append(review_visite)

    return(pd.DataFrame(rev_dict))
```

```python
#scrollable_popup = driver.find_element(By.XPATH, '/html/body/span[2]/g-lightbox/div/div[2]/div[3]/span/div/div/div/div[2]')

result_df = pd.DataFrame(columns = ['Review ID', 'Review Rate', 'Review Time','Review Visite','Review Text'
    ,'Review Voyage Type', 'Review Room','Review Emplacement', 'Review Service'])

# Normalement on le calcul avec nb_reviews sachant que par refresh on a 10 commentaires 
#nb_refresh = nb_reviews/10
#nb_refresh = 458
nb_refresh = 2000
refresh_time = 3 # A ajuster selon la rapidité de refresh, parce que s'il est trop court on saute des commentaires

for i in range(nb_refresh):
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    refresh_time = random.randrange(3,8)*0.1
    time.sleep(refresh_time)
    rnum = -random.randrange(10,50)
    driver.execute_script("window.scrollBy(0, %s);" % rnum)
    #time.sleep(1)
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    

    #time.sleep(refresh_time)
    if (i%5 == 0): print('Step '+str(i)+' - Scraping at '+str(round((i/nb_refresh)*100,2))+'%')
    if i%10 == 0 or i == nb_reviews:
        # Récupère tous les commentaires
        response = BeautifulSoup(driver.page_source, 'html.parser')

        r = response.find_all('div', class_="Svr5cf bKhjM")
        
        df = get_review_to_df(r)
        result_df = pd.concat([result_df, df], axis=0,
            join="outer",
            ignore_index=True,
            keys=None,
            levels=None,
            names=None,
            verify_integrity=False,
            copy=True,)
        print(result_df.shape)
```

```python
def convert_date(column):
    mois = {'janvier': '01','février': '02','mars':'03','avril':'04','mai':'05','juin':'06','juillet':'07','août':'08','septembre':'09','octobre':'10','novembre':'11','décembre':'12'}
    visite = list()

    for x in column:
        if x == 'no visite mentionned':
            a = None
        else :
            a = x[10:]
            if a[-1].isdigit()==False:
                a = a + ' 2023'
            a = mois[a[:-5]]+'/'+a[-4:]
        visite.append(a)
    return visite

```
